```python
# B번

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
csv_path = "./data/NFLX.csv"

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
# 0 - Data / 1 - Open / 2 - Hight / 3 - Low / 4 - Close / 5 - Adj_Close / 6 - Volume
df = pd.read_csv(csv_path)

df = pd.DataFrame(df)

# pd.to_datetime 을 이용하여 Date 타입을 바꿔준다
df["Date"] = pd.to_datetime(df["Date"])
# 조건을 통해서 2021 년 이후 데이터를 넣어준다
df = df[df["Date"] >= "2021-01-01"]

# DataFrame 출력
df
# plot(x축, y축, 선의 이름)
plt.plot(df["Date"], df["Close"], label="Close Price")
# 그래프 제목
plt.title("NFLX Close Price")
# 그래프 x축 이름
plt.xlabel("Date")
# 그래프 y축 이름
plt.ylabel("Close Price")
# 화면에 표시
plt.show()
```

```python
# D번

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
csv_path = "./data/NFLX.csv"

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
# 0 - Data / 1 - Open / 2 - Hight / 3 - Low / 4 - Close / 5 - Adj_Close / 6 - Volume
df = pd.read_csv(csv_path)

df = pd.DataFrame(df)

# pd.to_datetime 을 이용하여 Date 타입을 바꿔준다
df["Date"] = pd.to_datetime(df["Date"])
# 조건을 통해서 2021 년 이후 데이터를 넣어준다
df = df[df["Date"] >= "2021-01-01"]
# 년-월 로 데이터를 새로 생성함
df["year-month"] = df["Date"].dt.strftime("%Y-%m")
# 년-월로 만든 데이터를 기준으로 Close의 평균을 구함
answer = df.groupby("year-month")["Close"].mean()

# plot 결과를 그래프로 나타냄
plt.plot(answer, label="Avg Close Price")
# 그래프 제목
plt.title("Monthly Average Close Price")
# 그래프 x축 이름
plt.xlabel("Date")
# 그래프 y축 이름
plt.ylabel("Average Close Price")
# 화면에 표시
plt.show()
```

```python
# E번

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
csv_path = "./data/NFLX.csv"

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
# 0 - Data / 1 - Open / 2 - Hight / 3 - Low / 4 - Close / 5 - Adj_Close / 6 - Volume
df = pd.read_csv(csv_path)

df = pd.DataFrame(df)

# pd.to_datetime 을 이용하여 Date 타입을 바꿔준다
df["Date"] = pd.to_datetime(df["Date"])
# 조건을 통해서 2021 년 이후 데이터를 넣어준다
df = df[df["Date"] >= "2022-01-01"]
# 년-월 로 데이터를 새로 생성함

plt.plot(df["Date"],df["High"], label="High")
plt.plot(df["Date"],df["Low"], label="Low")
plt.plot(df["Date"],df["Close"], label="Close")
# 그래프 제목
plt.title("High, Low, and Close Price since January 2022")
# 그래프 x축 이름
plt.xlabel("Date")
# 그래프 y축 이름
plt.ylabel("Price")
plt.legend()
# 화면에 표시
plt.show()
```
